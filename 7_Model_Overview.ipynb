```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from scipy import stats

plt.rcParams["figure.figsize"] = (20, 15)
pd.set_option('display.max_columns', None)
```

```python
model_preformance = pd.read_csv('data\ModelPreformance.csv').sort_values('test_rmse').reset_index().drop(columns=['index'])
```

```python
model_preformance
```

```python
sns.set(font_scale=1.2)
plt.rcParams["figure.figsize"] = (18, 10)

values_melted = pd.melt(model_preformance[0:9], id_vars = 'model', 
                        value_vars = ['test_rmse', 'train_rmse'],
                        var_name='data', value_name = 'values')
ax = sns.barplot(data = values_melted, x = 'values', y = 'model', 
            hue = 'data', orient='h', palette = 'Paired')
ax.bar_label(ax.containers[0])
ax.set(title='Model Comparison, Train and Test RMSE')
ax.set_xlim(10000, 18500)
#sns.despine(left=True, bottom=True)
```
